```python
import math
from matplotlib import pyplot as plt
import numpy as np
```

```python
#TIME_QUANTA 
f = 3
dt = 1/(f*1000)
tbig = 0
tend = 1
n = int(tend/dt)
t = np.linspace(0,tend,n)
```

```python
#CREATING_SINEWAVE_INPUT 
Vm = 220
w = 2*math.pi*f
Vin = np.ones(np.size(t))
for i in range(np.size(t)):
    Vin[i] = 220*math.sin(w*t[i])
```

```python
plt.plot(t,Vin)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('Input Voltage')
```

```python
def halfWave(x):
    Vo = np.zeros(np.size(x))
    for i in range(np.size(x)):
        if x[i]>0:
            Vo[i] = x[i]
    return Vo
    
```

```python
Vo = halfWave(Vin)
```

```python
plt.plot(t,Vo)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('Output Voltage')
```

```python
def integrate(x,dx,n):
    S = 0
    for i in range(n-1):
        S += (x[i]+x[i+1])/2
    sum_term = S*dx
    return sum_term
```

```python
val = integrate(Vo*Vo,dt,n)
```

```python
print(np.sum(Vo*dt))
print(val**0.5)
```

```python
# Resistance LOAD 
R = 10
Io = Vo/R
```

```python
plt.plot(t,Io)
plt.plot(t,Vo,alpha=0.3)
plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.title('Output current with R load')
```

```python
# R L load 
R = 1
L = 1e-3
```

```python
def get_i(r,l):
    '''
    Analytic aproach to solve the problem
    '''
    z = (r**2 + (w*l)**2)**0.5 # defining impedence
    theta = math.atan((w*l)/r)
    tao = l/r
    A = Vm*math.sin(theta)/z
    I = np.zeros(np.size(t))
    temp_I = I
    beta = None
    def I_eqn(T):
        return (Vm/z)*np.sin(w*T-theta) + A*np.exp(-T/tao)
    for s in t:
        i = I_eqn(s)
        if (i == 0 and s != 0)  or (i<0):
            beta = s
            break
    
            
    
```

```python
I,beta = get_i(1,1)
```
